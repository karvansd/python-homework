```python
# Importing dependencies #

from pathlib import Path
import csv

# Declaring the location of the file #

filep = Path('budget/budget_data.csv')

# Initialize empty lists and variables to iterate through for the following variables: #

date = []                                   # Initialize the list variable to hold dates 
profit_losses = []                          # Initialize the list variable to hold Profit/Losses
monthly_profit_change = []                  # Initalize the list variable to hold monthly profit change
line_num = 0                                # Initialize the line number variable for budget_data.csv


# Open the input path as a file object in read mode #

with open(filep, 'r') as csvfile:

    # Storing the contents of the budget data as a csv file 
    csvreader = csv.reader(csvfile, delimiter=',')
    
    # Creating the header and line_num variables
    header = next(csvreader)
    line_num += 1
    
    # Iterating through to read each row of data after the header
    for row in csvreader:
    
        dates = row[0]                      # Set date variable equal to the value in the 1st column of each row
        date.append(dates)                  # Append the row salary value to the list of salaries
          
        profitloss = int(row[1])            # Set profit_losses variable equal to the value in the 2nd column of each row
        profit_losses.append(profitloss)    # Append the row profit/losses value to the list of profit and losses
        
        
# Initialize metric variables #
 
total_months = 0                            # Initialize the total number of months variables
net_total = 0                               # Initialize the net total of profit/losses variable
count_net_total = 0                         # Initialize the count of the net total profit/losses variable
average_profit_losses = 0                   # Initialize the average change in the profit/losses variable
greatest_increase_profits = 0               # Initialize the greatest increase in profits variable
greatest_decrease_profits = 0               # Initialize the greatest decrease in profits variable


#-------------------------------------#
# CALCULATING TOTAL MONTHS IN DATASET #
#-------------------------------------#

total_months = len(date)

print(f'There are {total_months} total months.')


#------------------------------------------------------------#
# CALCULATING NET TOTAL PROFIT/LOSSES OVER THE ENTIRE PERIOD #
#------------------------------------------------------------#

for num in profit_losses:
    net_total += num
    count_net_total += 1
    
print(f'The net profit and losses is {net_total}.')


#--------------------------------------------------------------------------------#
# CALCULATING THE AVERAGE IN THE CHANGES OF PROFIT/LOSSES OVER THE ENTIRE PERIOD #
#--------------------------------------------------------------------------------#

# Iterate through the profits in order to get the monthly change in profits
for i in range(len(profit_losses)-1):
        
    # Take the difference between two months and append to monthly profit change
    monthly_profit_change.append(profit_losses[i+1]-profit_losses[i])

average_profit_losses = round(sum(monthly_profit_change)/len(monthly_profit_change),2)

print(f'The average change in profit and losses over the entire period is {average_profit_losses}')

#-----------------------------------------------------------------#
# FINDING THE GREATEST INCREASE IN PROFITS OVER THE ENTIRE PERIOD #
#-----------------------------------------------------------------#

# The greatest increase in profits (date and amount) over the entire period.
max_increase_value = max(profit_losses)
max_month_index = profit_losses.index(max(profit_losses))
prev_month_amount = profit_losses[max_month_index - 1]
greatest_increase_profits = max_increase_value - prev_month_amount

# Finding the month of the greatest increase
greatest_increase_month = date[max_month_index]

print(f'The greatest increase in profits is ${greatest_increase_profits} on {greatest_increase_month}.')


#-----------------------------------------------------------------#
# FINDING THE GREATEST DECREASE IN PROFITS OVER THE ENTIRE PERIOD #
#-----------------------------------------------------------------#

# The greatest decrease in profits (date and amount) over the entire period.
min_increase_value = min(profit_losses)
min_month_index = profit_losses.index(min(profit_losses))
prev_month_min_amount = profit_losses[min_month_index - 1]
greatest_decrease_profits = min_increase_value - prev_month_min_amount

# Finding the month of the greatest decrease
greatest_decrease_month = date[min_month_index]

print(f'The greatest decrease in profits is ${greatest_decrease_profits} on {greatest_decrease_month}.')



# Output files #
output_file = Path("Financial_Analysis_Summary.txt")

with open(output_file,"w") as file:
    
# Write methods to print to Financial_Analysis_Summary #
    file.write("Financial Analysis")
    file.write("\n")
    file.write("----------------------------")
    file.write("\n")
    file.write(f"Total Months: {total_months}")
    file.write("\n")
    file.write(f"Total: ${net_total}")
    file.write("\n")
    file.write(f"Average Change: {average_profit_losses}")
    file.write("\n")
    file.write(f"Greatest Increase in Profits: {greatest_increase_month} ${greatest_increase_profits}")
    file.write("\n")
    file.write(f"Greatest Decrease in Profits: {greatest_decrease_month} ${greatest_decrease_profits}")
    
    
```
